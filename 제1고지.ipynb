```python
#-클래스 속성과 클래스 메서드
class Person:
  count = 0 #- 클래스 속성

  def __init__(self, data):
    Person.count += 1 #- 인스턴스가 만들어질 때, 클래스 속성에 1을 더함
    self.cnt = data #- 인스턴스 속성에 1 할당  

  @classmethod #- 클래스 메서드 만들기. 데코레이터(decorator)
  def print_count(cls): #- 클래스 메서드의 첫번째 매개변수는 cls
    print("{0}명 생성되었습니다.".format(cls.count)) #- cls.클래스속성 으로 클래스 속성에 접근 
    
    

jeongin = Person(10)
Person.print_count() #- 클래스 속성에 접근
heungseon = Person(9)
Person.print_count() #- 클래스 속성에 접근
huemin = Person(7)
Person.print_count() #- 클래스 속성에 접근

print(jeongin.cnt) #- 인스턴스 속성에 접근. 인스턴스.인스턴스속성
print(heungseon.cnt) #- 인스턴스 속성에 접근
print(huemin.cnt) #- 인스턴스 속성에 접근

jeongin.print_count
```

```python
#- step5.역전파 이론
import numpy as np


class Variable:
    def __init__(self, data):
        self.data = data


class Function:
    def __call__(self, input):
        x = input.data
        y = self.forward(x) #- 자식클래스의 인스턴스 메서트 foward()의 두번째 인자로 x를 입력한 뒤 출력값을 y에 할당.
        output = Variable(y)
        self.input = input
        #- 왜 self.input과 self.ouput에 할당하는 걸까 ? 
        #- class Function은 Square, Exp 등등의 특정한 기능을 하는 클래스의 부모 클래스로 활용할 것. 따라서 실제로 self는 자식클래스의 인스턴스를 가리킬 것. 
        #- 자식 클래스의 인스턴스에 인스턴스 속성 input과 ouput 값을 저장하는 것.  
        self.output = output
        return output

    def forward(self, x):
        raise NotImplementedError()


class Square(Function):
    def forward(self, x):
        return x ** 2


class Exp(Function):
    def forward(self, x):
        return np.exp(x)


def numerical_diff(f, x, eps=1e-4): 
  #- 수치미분 함수
  ''' << docstring
  Args:
    f: function 
    x: variable
  Returns:
    scalar 
  '''

  #- eps는 default 10의 -4제곱으로 작은 수를 입력. 
  x0 = Variable(x.data - eps) #- variable 인스턴스
  x1 = Variable(x.data + eps) 
  y0 = f(x0) #- f는 어떤 작용을 하는 함수. 입력값 x0을 받아 작용을 한 뒤 출력값을 y0에 할당.
  y1 = f(x1) #- 입력값 x1을 받아 작용을 한 뒤 출력값을 y1에 할당.
  return (y1.data - y0.data) / (2 * eps) #- 함수 f에 대해 x 지점에서의 수치 미분값


f = Square() #- f는 Square 클래스의 인스턴스.
x = Variable(np.array(2.0)) #- x는 Variable 클래스의 인스턴스이며, x.data를 np.array(2.0)로 초기화.
dy = numerical_diff(f, x) #- numerical_diff 함수의 인자로 f, x를 입력. 
print(dy) #- y= x^2 이라는 함수에서 x=2에 대한 수치적 미분 값


def f(x): #- 생소, 신선!! 
    A = Square()
    B = Exp()
    C = Square()
    return C(B(A(x))) #- 실행순서 x가 입력->A->B->C


x = Variable(np.array(0.5))
dy = numerical_diff(f, x)
print(dy) #- y = exp(x^2)^2 이라는 함수에서 x=0.5에 대한 수치적 미분 값
```

```python
#- step6.수동 역전파
import numpy as np


class Variable:
    def __init__(self, data):
        self.data = data #- data를 입력받아 초기화
        self.grad = None #- 미분값을 저장할건데 None으로 초기화  


class Function:
    def __call__(self, input):
        x = input.data
        y = self.forward(x)
        output = Variable(y)
        self.input = input #- self는 자기자신.
        #- input 과 self.input의 차이? 
        #- input은 Variable클래스의 인스턴스(전제)
        #- self.input은 인스턴스 속성. 우리는 Function을 다른 클래스의 기반 클래스로 활용할 것. 
        #- 예를들어 class Square(Function): 이때  f=Square()(x) 라면, self.input에는 x가 할당되며, 여기서 self는 Square 클래스의 인스턴스인 f 이다.  
        #- 클래스 변수 : 클래스 변수를 쓰면 클래스 내의 다른 메서드에서 사용이 가능하다. 한 클래스 내의 모~든 인스턴스들이 공통적인 값을 가져야 할때 클래스 변수로 선언. << 클래스가 로딩될 때 생성(메모리 할당)
        #- 인스턴스 변수 : 각각의 인스턴스마다 고유의 값을 가져야 할때 인스턴스 변수로 선언. << 인스턴스가 생성될 때 생성(메모리 할당) 
        #- 지역변수 : 메서드 내에서 선언되며 메서드 내에서만 사용할 수 있는 변수. 메서드가 실행될 때 메모리를 할당 받고, 메서드가 끝나면 소멸되어 사용 불가.  <=> 전역변수  
        return output

    def forward(self, x):
        raise NotImplementedError()

    def backward(self, gy):
        raise NotImplementedError()


class Square(Function):
    def forward(self, x):
        y = x ** 2 #- y는 메서드 내에서 선언되며 메서드 내에서만 사용할수 있는 지역변수. 굳이 y에 x**2값을 할당하여 return 하는 이유는~ 독자가 이해하기 쉽도록 
        return y

    def backward(self, gy):
        x = self.input.data 
        #- self는 Square 클래스의 인스턴스. 
        #- self.input은 "Square"(Function이 아님)의 인스턴스 변수, Sqaure의 인스턴스 변수 input은 Variable 클래스의 인스턴스. input.data는 input의 인스턴스 속성.
        #- 인스턴스의 순전파 입력값을 역전파에서 사용. 
        gx = 2 * x * gy #- gy의 의미 현재노드에서의 역전파 출력값은 local gradient 곱하기 global(upstream) gradient
        return gx


class Exp(Function):
    def forward(self, x):
        y = np.exp(x)
        return y

    def backward(self, gy):
        x = self.input.data
        gx = np.exp(x) * gy
        return gx


A = Square() #- Square 클래스의 인스턴스 A 생성(호출 아님)
B = Exp() #- Exp 클래스의 인스턴스 B 생성(호출 아님)
C = Square() #- Square 클래스의 인스턴스 C 생성(호출 아님)

#- 합성함수의 계산 그래프를 떠올리세요! 여기서 함성함수는 y = exp(x^2)^2 

#- 계산그래프에서 순전파를 떠올리세요.
#- x(입력) => A(square) => B(exponential) => C(square) => y(출력)
x = Variable(np.array(0.5))
a = A(x) #- x를 인자로 주면서 "Square 인스턴스 A가 호출"되면서 "Square의 부모클래스 Function의 __call__ 메서드가 실행"됨. a는 Variable 인스턴스. 
#- 왜 Funaction __call__ 메서드에서 output을 Variable 인스턴스로 내보내나? 
#- 딥러닝 모델에서 우리는 합성함수를 구현하고 싶음. y = C(B(A))에서 함수 A의 출력은 B의 입력이 되고, B의 출력은 C의 입력이 됨.
#- 우리는 함수를 구현할 때 Function class라는 함수의 기반클래스(함수들의 공통된 기능을 가짐)를 정의하고, 각 함수에서 Function class를 호출하여 사용함. 
#- 이때 Function __call__ 에서 Variable 인스턴스를 입력으로 받는다면, 출력도 Variable이어야 "합성함수" 구현 시 에러가 발생하지 않음. a.data = a (즉 x^2) a.grad=None
b = B(a) #- a를 인자로 주면서 "Exp 인스턴스 A가 호출"되면서 "Exp의 부모클래스 Function의 __call__ 메서드가 실행"됨. b는 Variable 인스턴스. b.data = b (즉 exp(x^2)) b.grad=None
y = C(b) #- b를 인자로 주면서 "Square 인스턴스 A가 호출"되면서 "Square의 부모클래스 Function의 __call__ 메서드가 실행"됨. y는 Variable 인스턴스. y.data= y (최종 출력값 exp(x^2)^2) y.grad=None

#- 계산그래프에서 역전파를 떠올리세요. 
#- y(출력) => C(square) => B(exponential) => A(square) => x(입력) 
#- 우리의 최종 목표는, dy/dx. 딥러닝 모델에서는 dL/dw 이겠죠? 역전파를 하는 이유는, Loss(손실,L)를 최소화하는 방향으로 Weight(가중치,W)를 갱신하기 위함이죠. 
y.grad = np.array(1.0) #- 인스턴스 변수 y.grad에 1.0 할당. y는 Variable 인스턴스이기 때문에 인스턴스 속성으로 y.data 와 y.grad가 있다. y.grad는 인스턴스 생성 시 None으로 초기화 되어있음. 
#- Square, Exp class 는 backward 메서드를 가지고 있음.  
b.grad = C.backward(y.grad) #- 계산그래프에서 C(Square) 노드를 역전파. Square 클래스의 backward 메서드는 upstream gradient를 입력받는다. 
#- 계산그래프에서 해당 노드의 역전파 출력값은  local gradient *(곱하기) global(upstream) gradient 이다. 
#- 따라서 b.grad = 2 * b * y.grad(1.0) << b는 C노드 순전파시 입력값 // local gradient(2b) * global gradient(1.0)
a.grad = B.backward(b.grad) #- 계산그래프에서 B(Exp) 노드를 역전파. 입력 인자는 upstream gradient, 즉 y(출력) => C(square) 역전파 출력값인 b.grad이다. 
#- 따라서 a.grad = exp(a)* b.grad << a는 B노드 순전파시 입력값 // local gradient(exp(a)) * global gradient(b.grad)
x.grad = A.backward(a.grad) #- 계산그래프에서 A(Square) 노드를 역전파. 입력 인자는 입력 인자는 upstream gradient, 즉 y(출력) => C(square) => B(exp) 역전파 출력값인 a.grad이다
#- 따라서 x.grad = 2 * x * a.grad << x는 A노드 순전파시 입력값 // local gradient(2x) * global gradient(a.grad)
print(x.grad)
```

```python
#- step7.역전파 자동화 
#- 문제제기: 계산을 할때마다 역전파 계산 코드를 수동으로 조합해야 하나... ? 
#- 위에서는 순전파가 x=>A=>B=>C=>y 라면 역전파는 y=>C=>B=>A=>x 를 "순서"에 따라 "직접 실행(수동)"했다.
#- 순전파를 한번만 하면, 역전파가 자동으로 이뤄지는 구조를 만들어보자. 
#- Define-by-Run 동적계산그래프
#- 함수의 순서를 리스트 형태로 저장한다면 ? => 웬거트 리스트 구조 ? 

#- 변수와 함수의 관계?
#- 함수에게 변수는 입력변수와 출력 변수로서 존재
#- 입력 변수 => 함수(창조자) => 출력변수  
#- 변수 관점에서 함수는, 창조자 creator 임 
#- 함수와 변수의 관계를 코드에 녹여보자!! 

import numpy as np


class Variable:
    def __init__(self, data):
        self.data = data
        self.grad = None
        self.creator = None #- creator라는 인스턴스 변수 추가

    def set_creator(self, func): #- set_creator라는 인스턴스 메서드 추가 
        self.creator = func #- 인스턴스 변수 self.creator에 입력받은 인자 func(함수)을 할당함.

    def backward(self):
        f = self.creator  # 1. Get a function. f에 함수를 할당 (self.creator = func)
        if f is not None: #- 만약 set_creator 인스턴스 메서드가 호출되어 self.creator 인스턴스 변수가 입력받은 인자 func로 갱신되었다면,
            x = f.input  # 2. Get the function's input. 함수의 input 
            x.grad = f.backward(self.grad)  # 3. Call the function's backward. Variable 인스턴스의 grad 값을 넣어 
            x.backward()

#값이 같은지 확인하려면 == 을 사용하시면 되고,
#참조가 같은지를 확인하려면 is를 사용하시면 됩니다.
#좀 더 자세히 정리하자면
#값과 참조 둘 다 같은지를 알고 싶다면 is를 사용하고.
#(참조가 같으면 같은 객체를 가리킬 것이고 당연히 값이 같겠죠?)
#참조는 달라도 되고 값만 비교하고 싶다. 하실 때는 ==, != 을 사용하면 됩니다.


class Function:
    def __call__(self, input):
        x = input.data
        y = self.forward(x) #- x를 인자로 하여 인스턴스의 순전파 메서드 호출.
        output = Variable(y)
        output.set_creator(self)  # Set parent(function) # 인스턴스가 set_creator의 func으로 들어감.
        #- set_creator(self, func)로 인자를 두개 입력받는데, 두번째 인자 func에 인스턴스가 입력됨.
        self.input = input #- 인스턴스 변수 input 에 인스턴스 호출 시 입력받은 인자 input을 할당 
        self.output = output  # Set output. 인스턴스 변수 ouput에 ouput 값 할당. 
        return output

    def forward(self, x):
        raise NotImplementedError()

    def backward(self, gy):
        raise NotImplementedError()


class Square(Function):
    def forward(self, x):
        y = x ** 2
        return y

    def backward(self, gy):
        x = self.input.data
        gx = 2 * x * gy
        return gx


class Exp(Function):
    def forward(self, x):
        y = np.exp(x)
        return y

    def backward(self, gy):
        x = self.input.data
        gx = np.exp(x) * gy
        return gx


A = Square()
B = Exp()
C = Square()

x = Variable(np.array(0.5))
a = A(x) 
#- output.set_creator(self) << 여기서 self는 인스턴스 자기자신
#- output.creator = A 
#- A.input = input
#- A.output = output
b = B(a)
#- output.set_creator(self) << 여기서 self는 인스턴스 자기자신
#- output.creator = B 
#- B.input = input
#- B.output = output
y = C(b)
#- output.set_creator(self) << 여기서 self는 인스턴스 자기자신
#- output.creator = C 
#- C.input = input
#- C.output = output
#- y는 최종출력값을 인자로 넣은 Variable 인스턴스  

# backward 자동화 완성! 
y.grad = np.array(1.0) 
y.backward() # 이 backward는 Variable의 인스턴스 메서드이다.
#- 1. f = C << 이 시점에서 self.creator에는 C가 저장됨. 
#- C는 Square()클래스의 인스턴스 
#- 2. f는 None 값이 아니기에 조건문 실행(1회)
#- 3. x에 C.input 인 input 값을 할당함. x == f.input == C.input == b == B(a)
#- 4. C.backward()의 인자로 self.grad(global gradient)를 넣고 역전파 출력값을 x.grad에 반환.
#- 이후 x.backward()를 재귀적으로 진행! 
#- 6. x = B(a)이므로, f = B
#- 7. f는 None 값이 아니기에 조건문 실행(2회)
#- 8. x에 B.input 인 input 값을 할당. x == f.input == B.input == a == A(x)
#- 9. B.backward()의 인자로 self.grad(global gradient)를 넣고 역전파 출력값을 x.grad에 반환.
#- 6. x = A(b)이므로, f = A
#- 7. f는 None 값이 아니기에 조건문 실행(3회)
#- 8. x에 A.input 인 input 값을 할당. x == f.input == A.input == x == Variable(np.array(0.5))
#- 9. A.backward()의 인자로 self.grad(global gradient)를 넣고 역전파 출력값을 x.grad에 반환.
#- 10. x = Variable(np.array(0.5)) 이므로, f = None
#- 재귀문은 f가 None이 되면 끝남. 
print(x.grad)

'''
def backward(self):
    f = self.creator  # 1. Get a function. f에 함수를 할당 (self.creator = func)
    if f is not None: #- 만약 set_creator 인스턴스 메서드가 호출되어 self.creator 인스턴스 변수가 입력받은 인자 func로 갱신되었다면,
        x = f.input  # 2. Get the function's input. 함수의 input 
        x.grad = f.backward(self.grad)  # 3. Call the function's backward. Variable 인스턴스의 grad 값을 넣어 
        x.backward()
'''
```

```python
#- step9.함수를 더 편리하게

import numpy as np


class Variable:
    def __init__(self, data):
        if data is not None:
            if not isinstance(data, np.ndarray): #- 인자로 입력받은 data가 np.ndarray인스턴스인지 확인. DeZero의 Variable은 데이터로 ndarray 인스턴스만 취급하고 싶다.
                raise TypeError('{} is not supported'.format(type(data))) #- Variable이 ndarray 외의 데이터를 넣을 경우, 오류 발생. 

        self.data = data
        self.grad = None
        self.creator = None

    def set_creator(self, func):
        self.creator = func

    def backward(self): #- 역전파 자동화를 반복문으로 구현 
        if self.grad is None: #- 인스턴스 변수 grad 값을 입력하지 않으면 
            self.grad = np.ones_like(self.data) #- 자동으로 미분값을 생성(1로 초기화).
            #- np.ones_like(self, data)는 self.data와 형상 및 데이터 타입이 같은 ndarray 인스턴스를 생성하는데, 모든 요소를 1로 채워 돌려줌. 

        funcs = [self.creator]
        while funcs:
            f = funcs.pop()
            x, y = f.input, f.output
            x.grad = f.backward(y.grad)

            if x.creator is not None:
                funcs.append(x.creator)


def as_array(x):
    if np.isscalar(x): #- np.isscalar 는 입력 데이터가 numpy.float64 같은 스칼라 타입인지 확인.
        return np.array(x) #- 입력이 스칼라인 경우, ndarray인스턴스로 변환
    return x


class Function:
    def __call__(self, input):
        x = input.data
        y = self.forward(x)
        output = Variable(as_array(y)) #- 출력값 y에 as_array 함수를 적용한뒤 Variable로 감쌈 
        #- 1.반환값 out은 Variable instance이고, 2.output.data는 ndarray instance 이다.
        output.set_creator(self)
        self.input = input
        self.output = output
        return output

    def forward(self, x):
        raise NotImplementedError()

    def backward(self, gy):
        raise NotImplementedError()


class Square(Function):
    def forward(self, x):
        y = x ** 2
        return y

    def backward(self, gy):
        x = self.input.data
        gx = 2 * x * gy
        return gx


class Exp(Function):
    def forward(self, x):
        y = np.exp(x)
        return y

    def backward(self, gy):
        x = self.input.data
        gx = np.exp(x) * gy
        return gx


def square(x): #- 파이썬 함수 구현 
    return Square()(x)


def exp(x):
    return Exp()(x)


x = Variable(np.array(0.5)) #- x를 Variable로 감싸면 함수를 연속으로 적용 가능.
y = square(exp(square(x))) #- 함수를 연속하여 적용
#- y.grad = np.array(1.0) 
y.backward() 
print(x.grad)


x = Variable(np.array(1.0))  # OK
x = Variable(None)  # OK
x = Variable(1.0)  #- TypeError('{} is not supported'.format(type(data))) => <class 'float'> is not supported
```

```python
#- step10.테스트
#- 테스트를 통해 실수(버그)를 예방할수 있음.
#- 테스트를 자동화해야 소프트웨어의 품질을 유지할 수 있음. 
#- 딥러닝 프레임워크의 간단~한 테스트 방법을 살펴보자.

import unittest #- 파이썬으로 테스트하는 경우, 표준 라이브러리에 포함된 unittest 사용하면 편리함
import numpy as np


class Variable:
    def __init__(self, data):
        if data is not None:
            if not isinstance(data, np.ndarray):
                raise TypeError('{} is not supported'.format(type(data)))

        self.data = data
        self.grad = None
        self.creator = None

    def set_creator(self, func):
        self.creator = func

    def backward(self):
        if self.grad is None:
            self.grad = np.ones_like(self.data)

        funcs = [self.creator]
        while funcs:
            f = funcs.pop()
            x, y = f.input, f.output
            x.grad = f.backward(y.grad)

            if x.creator is not None:
                funcs.append(x.creator)


def as_array(x):
    if np.isscalar(x):
        return np.array(x)
    return x


class Function:
    def __call__(self, input):
        x = input.data
        y = self.forward(x)
        output = Variable(as_array(y))
        output.set_creator(self)
        self.input = input
        self.output = output
        return output

    def forward(self, x):
        raise NotImplementedError()

    def backward(self, gy):
        raise NotImplementedError()


class Square(Function):
    def forward(self, x):
        y = x ** 2
        return y

    def backward(self, gy):
        x = self.input.data
        gx = 2 * x * gy
        return gx


def square(x):
    return Square()(x)

    # 어떤 수식을 사용하려면 
    #1. Function을 만든다.
    #2. Function을 상속받는 구체적인 계산 class를 만든다.
    #3. 2를 파이썬 함수 형태로 만든다.


def numerical_diff(f, x, eps=1e-4):
    x0 = Variable(x.data - eps)
    x1 = Variable(x.data + eps)
    y0 = f(x0)
    y1 = f(x1)
    return (y1.data - y0.data) / (2 * eps)


class SquareTest(unittest.TestCase): #- unittest.TestCase를 상속한 SquareTest클래스를 구현.
#- 메서드 이름규칙 => test_테스트내용
#- 기대되는 결과를 확인하기 위해 assertEqual()를 호출 
#- 조건을 검증하기 위해 assertTrue() 또는 assertFalse()를 호출 
#- 특정 예외가 발생했는지 검증하기 위해 assertRaises()를 호출

    def test_forward(self): #- square함수의 순전파 테스트. 출력값이 기댓값과 같은지 확인
        x = Variable(np.array(2.0))
        y = square(x)
        expected = np.array(4.0)
        self.assertEqual(y.data, expected)

    def test_backward(self): #- square함수의 역전파 테스트
        x = Variable(np.array(3.0))
        y = square(x)
        y.backward()
        expected = np.array(6.0)
        self.assertEqual(x.grad, expected)

    def test_gradient_check(self):
        x = Variable(np.random.rand(1))
        y = square(x)
        y.backward()
        num_grad = numerical_diff(square, x)
        flg = np.allclose(x.grad, num_grad) #- 수치미분과 역전파 결과값 비교
        #- np.allclose(a,b)는 ndarray 인스턴스인 a와 b의 값이 가까운지 판정. 
        self.assertTrue(flg) #- flg 조건 검증 
```
